```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data.csv")
data.head()
```

```python
data['split'] = np.random.randn(data.shape[0], 1)

msk = np.random.rand(len(data)) <= 0.8

train = data[msk]
test = data[~msk]
```

```python
print(train.shape)
print(test.shape)
train.head()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

columns = ["diagnosis"]
for col in columns:
  train[col] = le.fit_transform(train[col])
  test[col] = le.transform(test[col])
  print(le.classes_)

train.head()
test_ids = test["id"]
```

```python
train.head()
```

```python
train = train.drop(["id", "Unnamed: 32", "split"], axis=1)
test = test.drop(["id", "Unnamed: 32", "split"], axis=1)
```

```python
Y = train["diagnosis"]
X = train.drop(["diagnosis"], axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(x_train, y_train)
```

```python
predictions = rf.predict(x_test)
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, predictions)
```

```python
Y_new = test["diagnosis"]
Xnewtest = test.drop(["diagnosis"], axis=1)
final_predictions = rf.predict(Xnewtest)
accuracy_score(Y_new, final_predictions)
```

```python
df = pd.DataFrame({"IDS": test_ids.values,
                   "Original Results": Y_new.values,
                   "Predictions": final_predictions,})
```

```python
df.to_csv("RandomForest_predictions.csv", index=False)
```
