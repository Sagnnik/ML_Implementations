```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data.csv")
```

```python
data.head()
```

```python
data = data.drop("Unnamed: 32", axis=1)
```

```python
data.head()
```

```python
data['split'] = np.random.randn(data.shape[0], 1)

msk = np.random.rand(len(data)) <= 0.8

train = data[msk]
test = data[~msk]
```

```python
print(train.shape)
print(test.shape)
train.head()
```

```python
test_ids = test["id"]
train = train.drop(["split","id"], axis=1)
test = test.drop(["split","id"], axis=1)
```

```python
train.head()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

columns = ["diagnosis"]
for col in columns:
  train[col] = le.fit_transform(train[col])
  test[col] = le.transform(test[col])
  print(le.classes_)

train.head()
```

```python
from sklearn.model_selection import train_test_split

Y = train["diagnosis"]
X = train.drop(["diagnosis"], axis=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

logModel = LogisticRegression(random_state=0, max_iter=5000)
```

```python
logModel.fit(x_train, y_train)
```

```python
predictions = logModel.predict(x_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predictions)
```

```python
test.head()
```

```python
new_column = test["diagnosis"]
test = test.drop(["diagnosis"], axis=1)
test.head()
```

```python
final_predictions = logModel.predict(test)
```

```python
accuracy_score(new_column, final_predictions)
```

```python
df = pd.DataFrame({"PatientID": test_ids.values,
                   "Original Data": new_column.values,
                   "Predicions": final_predictions,
                  })
```

```python
df.to_csv("Prediction_list_Logistic.csv", index=False)
```
